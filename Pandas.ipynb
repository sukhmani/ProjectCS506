1. Introduction to Pandas
Pandas is an open-source Python library that provides high-performance, easy-to-use data structures and data analysis tools. It is widely used for data manipulation and analysis in Python. 

Key features of Pandas include:

- **Data Structures**: Pandas provides two primary data structures:
    - `Series`: A one-dimensional labeled array capable of holding any data type.
    - `DataFrame`: A two-dimensional labeled data structure, similar to a table in a relational database or an Excel spreadsheet.

- **Data Manipulation**: Pandas allows for operations such as filtering, grouping, merging, reshaping, and cleaning data.

- **File I/O**: It supports reading and writing data from various file formats, including CSV, Excel, JSON, SQL, and more.

- **Time Series**: Pandas has robust support for working with time-series data.

Pandas is built on top of NumPy and integrates well with other libraries in the Python ecosystem, such as Matplotlib and Scikit-learn. It is a powerful tool for data scientists and analysts.

```python
# Basic operations with Pandas DataFrame
import pandas as pd

# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
print(df)

df.head()  # Shows first five rows
df.tail()  # Shows last five rows
df.info()  # Summary of DataFrame structure
df.describe()  # Statistical summary of numerical columns

```

```python
#3) Selecting Data
print(df['Name'])  # Selects a single column
print(df[['Name', 'Age']])  # Selects multiple columns
print(df.iloc[0])  # Selects first row by index
print(df.loc[0])  # Selects first row using label
```

```python
#Filtering Data
filtered_df = df[df['Age'] > 28]  # Filter rows where Age > 28
print(filtered_df)
```

```python
#Modifying Data
df['Age'] = df['Age'] + 1  # Modify column values
df['City'] = ['NY', 'LA', 'SF']  # Add new column
df.rename(columns={'Age': 'Years'}, inplace=True)  # Rename column

```

```python
#Handling Missing Values
df.dropna()  # Remove rows with missing values
df.fillna(value=0)  # Replace missing values with 0

```

```python
import pandas as pd

df = pd.read_csv("GlobalWeatherRepository.csv")
print(df.head())  # Preview the first few rows
```

```python
# Check data structure
print(df.info())  # Column details and types
print(df.describe())  # Statistical summary
```

```python
#Filter Data
hot_days = df[df["temperature_celsius"] > 30]
print(hot_days)
```

```python
humid_places = df[df["humidity"] > 80]
print(humid_places)
```

```python
humid_places = df[df["humidity"] > 80]
print(humid_places)
```

```python
polluted_places = df[df["air_quality_PM2.5"] > 50]
print(polluted_places)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df["temperature_celsius"], bins=20)
plt.xlabel("Temperature (C)")
plt.ylabel("Frequency")
plt.title("Temperature Distribution Across Locations")
plt.show()
```

```python
#Visualizing Air Pollution Data You can create a bar chart or histogram to display pollution levels by location.
import matplotlib.pyplot as plt

df.plot(kind="bar", x="location_name", y="air_quality_PM2.5")
plt.title("PM2.5 Levels by Location")
plt.ylabel("PM2.5 Concentration")
plt.xticks(rotation=90)
plt.show()
```

```python
#Analyzing Pollution Trends If your dataset has time-based records, you can check how pollution fluctuates.
import seaborn as sns

sns.lineplot(x=df["last_updated"], y=df["air_quality_PM2.5"])
plt.xticks(rotation=45)
plt.title("PM2.5 Pollution Over Time")
plt.show()
```

```python
correlation = df[["air_quality_PM2.5", "air_quality_Ozone", "air_quality_Nitrogen_dioxide"]].corr()
print(correlation)
```
